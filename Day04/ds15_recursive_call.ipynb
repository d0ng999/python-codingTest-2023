### 재귀호출(Recursive Call)

```python
count = 10

def openBox():
    global count
    print('종이 상자를 엽니다.')
    count-=1
    if count ==0:
        print('반지를 넣고 반환합니다.')
        return
    # 리턴은 함수를 호출한 그 위치로 간다. 즉, 카운트가 0이 될 때까지 위의 함수를 반복
    # 하다가 10번째에 리턴을 만나면 9번째 리턴으로 넘어가서 남은 함수를 그대로 진행
    # 9번째 리턴은 8번째 리턴으로 올라가서 함수 진행 ....
    # 어렵다

    openBox()
    print('종이상자를 닫습니다.')
```

```python
count = 2
openBox()
```

### 재귀호출 하지 않고 반복문으로 계산

```python
sumValue = 0

for n in range(10, 0, -1):
    sumValue += n

print('10+9+8+...+1 = ', sumValue)
```

재귀호출로 덧셈

```python
def addNumber(num):
    if num <=1:
        return 1
    else:
        return num + addNumber(num-1)
    
print(addNumber(3))
```

팩토리얼 반복문

```python
factValue = 1
for n in range(50,0,-1):
    factValue *= n
print(factValue)
```

```python
def factorialNum(num):
    if num <=1:
        return 1
    else:
        return num * factorialNum(num-1)
    
print(factorialNum(1558))
```

우주선 발사

```python
import time
def countdown(n):
    if n == 0:
        print('우주선 발사')
    else:
        print(n)
        time.sleep(1)
        countdown(n-1)
countdown(10)
```

```python
def printStar(n):
    if n>0:
        print('☆'*n)
        printStar(n-1)
        
printStar(10)
```

구구단 재귀호출

```python
def gugudan(dan, num):
    print(f'{dan} x {num} = {dan*num:5.2f}', end = ' ')
    if num < 9:
        gugudan(dan, num+1)
for dan in range(2,10):
    print(f'\n{dan}단 시작', end = ' ')
    gugudan(dan,1)
```

N제곱

```python
tab = ''
def pow(x, n):
    global tab
    tab += ' '
    if n ==0:
        return 1
    print(tab, f'{x}x{x}^{n}-{1}')
    return x*pow(x,n-1)

print('2^10')
print('답 --> ', pow(2,10))
```

피보나치 수열

```python
def fibonacci(n):
    if n == 0: 
        return 0
    elif n ==1: 
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print('피보나치 수 --> 0 1 ', end = '')
for i in range(2, 30):
    print(fibonacci(i), end =' ')
```

회문 여부

```python
def palindrome(pStr):
    if len(pStr) <= 1:
        return True
    
    if pStr[0] != pStr[-1]:
        return False
    
    else:
        return palindrome(pStr[1:len(pStr)-1])
```

```python
strAry = ['reaver', 'kayak', 'Borrow or rob', '주유소의 소유주', '야 너 이번 주 주번이 너 야', '살금 살금']
```

```python
for testStr in strAry:
    print(testStr, end = '--> ')
    testStr = testStr.lower().replace(' ','') # replace문 공백을 없는 것으로 대체
    if palindrome(testStr) == True:
        print('회문입니다!.')
    else:
        print('회문 아니오--;')
```
